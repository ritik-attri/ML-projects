```python
happy_dataset=np.load('imgdataset\Happy.npy')
sad_dataset=np.load('imgdataset\Sad.npy')
#angry_dataset=np.load('imgdataset\Angry.npy')
print(happy_dataset.shape)
print(sad_dataset.shape)
#print(angry_dataset.shape)
```

```python
whole_dataset=np.concatenate((sad_dataset,happy_dataset))
```

```python
Y_dataset=np.zeros((2000,1))
Y_dataset[1000:,0]=1
Y_dataset[:1000,0]=0
print(Y_dataset.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(whole_dataset,Y_dataset,test_size=0.2,shuffle=True)
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
print(y_train.shape)
```

```python
face_cascade=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
cap=cv2.VideoCapture(0)
face_data=[]
i=0
predict=[]
while True:
    time.sleep(0.01)
    ret,frame=cap.read()
    if not ret:
        continue

    

    faces=face_cascade.detectMultiScale(frame,1.3,5)
    faces=sorted(faces,key=lambda x:x[2]*x[3],reverse=True)
    faces=faces[:1]
    for face in faces:
        x,y,w,h=face
        face_selection=frame[y:y+h,x:x+w]
        face_selection=cv2.resize(face_selection,(100,100))
        face_selection=np.array(face_selection)
        face_selection=face_selection.reshape(1,-1)
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),5)
        cv2.putText(frame,str(''),(x+w,y+h),cv2.FONT_HERSHEY_COMPLEX,0.9,(255,0,0),1)
        face_data.append(face_selection)
        face_data_array=np.array(face_data[i])
        face_data_array=face_data_array.reshape((1,100,100,3))
        print((model.predict(face_data_array.reshape(1,100,100,3))[0][0]))
        predict=model.predict(face_data_array.reshape(1,100,100,3))
        print(predict)
        if model.predict(face_data_array.reshape(1,100,100,3))[0][0]>model.predict(face_data_array.reshape(1,100,100,3))[0][1]:
            cv2.putText(frame,str('Sad'),(x,y),cv2.FONT_HERSHEY_COMPLEX,0.9,(255,0,0),1)
        else:
            cv2.putText(frame,str('Happy'),(x,y),cv2.FONT_HERSHEY_COMPLEX,0.9,(255,0,0),1)
        i+=1
       

    cv2.imshow('Feed',frame)
    key=cv2.waitKey(1)
    if key & 0xFF==ord('q'):
        break		

cap.release()
cv2.destroyAllWindows() 
```

```python
model=models.Sequential()

model.add(layers.Conv2D(3,(3,3),input_shape=(100,100,3)))
model.add(layers.Activation('relu'))
model.add(layers.MaxPool2D((2,2)))
    
model.add(layers.Conv2D(3,(3,3),input_shape=(100,100,3)))
model.add(layers.Activation('relu'))
model.add(layers.MaxPool2D((2,2)))

model.add(layers.Conv2D(3,(3,3),input_shape=(100,100,3)))
model.add(layers.Activation('relu'))
model.add(layers.MaxPool2D((2,2)))
    

    
model.add(layers.Flatten())
model.add(layers.Dense(100))
model.add(layers.Activation('sigmoid'))
model.add(layers.Dense(2))
model.add(layers.Activation('softmax'))

    
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```
